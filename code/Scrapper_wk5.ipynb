```python
import pandas as pd
import numpy as np
import os
from bs4 import BeautifulSoup
import json
import requests
import janitor
```

```python
url = "https://explodingtopics.com/blog/music-streaming-stats"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
table = pd.read_html(soup.prettify())
streaming = table[11]
```

```python
streaming.dropna(inplace=True)
streaming
```

```python
streaming.columns = streaming.iloc[0]
streaming = streaming[1:]
```

```python
def clean_value(value):
    if value == "-" or value == "" or pd.isna(value):
        return 0
    else:
        value = str(value).lower().strip()
        if 'million' in value:
            value = value.replace('million', '').strip()
            return float(value)

streaming = streaming.clean_names()

for col in streaming.columns:
    if col != "year":
        streaming[col] = streaming[col].apply(clean_value)

streaming["year"] = streaming["year"].astype(int)
streaming
```
