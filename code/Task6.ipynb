```python
import requests
import json
import pandas as pd
import time
import os
```

```python
BASE_URL = "https://unidbapi.com/api/degree/read_ratings"
API_KEY = "227681c508"
```

```python
locations = [
    {"name": "ComputerScience",
     "filename": "computer_science.json"},
    {"name": "Mathematics",
     "filename": "mathematics.json"},
    {"name": "History",
     "filename": "history.json"},
    {"name": "Biology",
     "filename": "biology.json"},
    {"name": "Law",
     "filename": "law.json"},
    {"name": "Architecture",
     "filename": "architecture.json"},
    ]
```

```python
for i, location in enumerate(locations, 1):
  params = {'d': location["name"],
            'key': API_KEY}
  response = requests.get(BASE_URL, params=params)
  data = response.json()
  data_with_metadata = {
            'location': location['name'],
            'total_unis': len(data),
            "ratings": data}
  with open(location['filename'], 'w') as f:
    json.dump(data_with_metadata, f, indent=2)
```
